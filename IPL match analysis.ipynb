# Part three:
1. Who has lots of sixes ?
2.  Top five strike rate(in death over ) players in IPL history. But one Condition is, every player has to be played atleast 200 balls.
3.  Height score in every IPL season.

```python
deli = pd.read_csv("deliveries.csv")
deli.head()
```

```python
mask = deli["batsman_runs"] == 6
deli[mask].groupby("batsman")["batsman_runs"].count().sort_values(ascending = False) # 1. So, Gayle has made 266 sixes.
```

```python
death_over = deli[deli["over"] > 15]
death_over.head()
```

```python
x = death_over.groupby("batsman")["batsman_runs"].count()
final_players = x[x > 200].index.tolist()
final_players
```

```python
final = death_over[death_over["batsman"].isin(final_players)]
balls = final.groupby("batsman")["batsman_runs"].count()
runs = final.groupby("batsman")["batsman_runs"].sum()
```

```python
sr = (runs / balls ) * 100
sr.sort_values(ascending = False).head()
```

```python
sr.sort_values(ascending = False).head().plot(kind ="bar") # Question 2 is done here.
```

```python
merge_set = deli.merge(df,left_on = "match_id",right_on = "id")
merge_set.head()
```
